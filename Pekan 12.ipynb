```python
import math

while True:
    try:
        jari_jari = input("Masukkan jari-jari lingkaran: ")
        if not jari_jari:  # Jika input kosong
            raise ValueError("tidak boleh kosong")
        jari_jari = float(jari_jari)  # Mengonversi input ke float
        break  # Keluar dari loop jika input valid
    except ValueError as e:
        if str(e) == "tidak boleh kosong":
            print("Input tidak boleh kosong.")
        else:
            print("Input hanya boleh memasukkan angka.")

while True:
    try:
        sisi = input("Masukkan sisi persegi: ")
        if not sisi:  # Jika input kosong
            raise ValueError("tidak boleh kosong")
        sisi = float(sisi)  # Mengonversi input ke float
        break  # Keluar dari loop jika input valid
    except ValueError as e:
        if str(e) == "tidak boleh kosong":
            print("Input tidak boleh kosong.")
        else:
            print("Input hanya boleh memasukkan angka.")

# Menghitung luas lingkaran
luas_lingkaran = math.pi * (jari_jari ** 2)
print(f"Luas lingkaran adalah {luas_lingkaran:.2f}")

# Menghitung luas persegi
luas_persegi = sisi * sisi
print(f"Luas persegi adalah {luas_persegi:.2f}")
```

```python
while True:
    try:
        nomor = input("Masukkan NPM Anda: ")
        if not nomor:  # Jika input kosong
            raise ValueError("tidak boleh kosong")
        if not nomor.isdigit():  # Jika input bukan angka
            raise ValueError("hanya boleh memasukkan angka")
        break  # Keluar dari loop jika input valid
    except ValueError as e:
        if str(e) == "tidak boleh kosong":
            print("Input tidak boleh kosong.")
        elif str(e) == "hanya boleh memasukkan angka":
            print("Input hanya boleh memasukkan angka.")

genap = 0
ganjil = 0

# Menghitung jumlah angka genap dan ganjil
for no in nomor:
    if int(no) % 2 == 0:
        genap += 1
    else:
        ganjil += 1

# Menampilkan hasil
print("Jumlah angka genap:", genap)
print("Jumlah angka ganjil:", ganjil)
```
