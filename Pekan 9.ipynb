```python
#DICTIONARY
mahasiswa = [
    {
        "nama": "Akmal Firdaus",
        "kelas": "1",
        "tanggal_lahir": "25-05-06",
        "no_hp": "081234567890",
        "alamat": "Gombong, Kebumen",
        "hobi": ["Membaca", "Bermain game", "Berenang"]
    },
]
```

```python
for mhs in mahasiswa:
    print("Nama:", mhs["nama"])
    print("Kelas:", mhs["kelas"])
    print("Tanggal Lahir:", mhs["tanggal_lahir"])
    print("No HP:", mhs["no_hp"])
    print("Alamat:", mhs["alamat"])
    print("Hobi:")
    for hobi in mhs["hobi"]:
        print("-", hobi)
```
