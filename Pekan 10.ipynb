```python
def faktorial(n):
  """Fungsi untuk menghitung faktorial secara rekursif.

  Args:
    n: Bilangan bulat positif yang akan dihitung faktorialnya.

  Returns:
    Nilai faktorial dari n.
  """

  if n == 0:
    return 1
  else:
    return n * faktorial(n-1)

bilangan = int(input("Masukkan bilangan: "))

hasil = faktorial(bilangan)
print(f"Faktorial dari {bilangan} adalah {hasil}")
```

```python
import math

def hitung_luas_lingkaran(jari_jari):
  """Fungsi untuk menghitung luas lingkaran.

  Args:
    jari_jari: Panjang jari-jari lingkaran.

  Returns:
    Luas lingkaran.
  """

  luas = math.pi * jari_jari * jari_jari
  return luas

def hitung_luas_persegi(sisi):
  """Fungsi untuk menghitung luas persegi.

  Args:
    sisi: Panjang sisi persegi.

  Returns:
    Luas persegi.
  """

  luas = sisi * sisi
  return luas

def hitung_luas_segitiga(alas, tinggi):
  """Fungsi untuk menghitung luas segitiga.

  Args:
    alas: Panjang alas segitiga.
    tinggi: Tinggi segitiga.

  Returns:
    Luas segitiga.
  """

  luas = 0.5 * alas * tinggi
  return luas

while True:
  print("Pilih bangun datar yang ingin dihitung luasnya:")
  print("1. Lingkaran")
  print("2. Persegi")
  print("3. Segitiga")
  print("4. Keluar")

  pilihan = input("Masukkan pilihan Anda (1/2/3/4): ")

  if pilihan == '1':
    jari_jari = float(input("Masukkan panjang jari-jari: "))
    luas = hitung_luas_lingkaran(jari_jari)
    print(f"Luas lingkaran adalah: {luas:.2f}")
  elif pilihan == '2':
    sisi = float(input("Masukkan panjang sisi: "))
    luas = hitung_luas_persegi(sisi)
    print(f"Luas persegi adalah: {luas:.2f}")
  elif pilihan == '3':
    alas = float(input("Masukkan panjang alas: "))
    tinggi = float(input("Masukkan tinggi: "))
    luas = hitung_luas_segitiga(alas, tinggi)
    print(f"Luas segitiga adalah: {luas:.2f}")
  elif pilihan == '4':
    print("Terima kasih telah menggunakan kalkulator bangun datar!")
    break
  else:
    print("Pilihan tidak valid. Silakan pilih 1, 2, 3, atau 4.")
```
