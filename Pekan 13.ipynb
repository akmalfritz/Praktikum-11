```python
class Kucing:
    def __init__(self, nama, warna):
        self.nama = nama       # Atribut
        self.warna = warna     # Atribut

    def meong(self):           # Metode
        return f"{self.nama} berwarna {self.warna} mengeong!"
```

```python
# Membuat objek dari class Kucing
kucing1 = Kucing("Tom", "abu-abu")
kucing2 = Kucing("Jerry", "putih")

# Mengakses metode objek
print(kucing1.meong())  # Output: Tom berwarna abu-abu mengeong!
print(kucing2.meong())  # Output: Jerry berwarna putih mengeong!
```

```python
class AlatTulis:
    def __init__(self, nama, stok, harga_satuan):
        self.nama = nama
        self.stok = stok
        self.harga_satuan = harga_satuan
        self.harga = self.stok * self.harga_satuan

    # Setter untuk memperbarui nilai nama, stok, harga_satuan, dan harga
    def set_data(self, nama, stok, harga_satuan):
        self.nama = nama
        self.stok = stok
        self.harga_satuan = harga_satuan
        self.harga = self.stok * self.harga_satuan

    # Getter untuk menampilkan nilai nama, stok, harga_satuan, dan harga
    def get_data(self):
        return {
            "Nama": self.nama,
            "Stok": self.stok,
            "Harga Satuan": self.harga_satuan,
            "Harga Total": self.harga,
        }


class Penjualan:
    def __init__(self):
        self.total_harga = 0

    # Setter untuk menghitung total harga penjualan
    def set_total_harga(self, daftar_alat_tulis):
        self.total_harga = sum(item.harga for item in daftar_alat_tulis)

    # Getter untuk menampilkan total harga penjualan
    def get_total_harga(self):
        return self.total_harga


# Data alat tulis
bolpoint = AlatTulis("Bolpoint", 10, 2000)
pensil = AlatTulis("Pensil", 10, 1000)
penghapus = AlatTulis("Penghapus", 10, 500)

# Menampilkan data alat tulis
print("Data Alat Tulis:")
for item in [bolpoint, pensil, penghapus]:
    print(item.get_data())

# Mengelola transaksi penjualan
penjualan = Penjualan()
penjualan.set_total_harga([bolpoint, pensil, penghapus])
print("\nTotal Harga Penjualan:", penjualan.get_total_harga())
```
